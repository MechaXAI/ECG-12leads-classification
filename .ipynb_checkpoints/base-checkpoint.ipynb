# ECG Signal Processing and Feature Extraction 

```python
import neurokit2 as nk
import numpy as np
import pandas as pd

# Limpieza y deteccin de R-peak.
processed_signals, info = nk.ecg_process(ecg_signal, sampling_rate=sampling_rate)
nk.ecg_plot(processed_signals, info)
```

```python
# Extract heart rate variability time-domain features based on the R-peaks.
rpeaks = processed_signals["ECG_R_Peaks"]
hrv_time = nk.hrv_time(rpeaks, sampling_rate=sampling_rate, show=False)
print(hrv_time.columns)
```

```python
features = {
        "HRV_MeanNN": hrv_time.get("HRV_MeanNN", np.nan).iloc[0],
        "HRV_SDNN": hrv_time.get("HRV_SDNN", np.nan).iloc[0],
        "HRV_RMSSD": hrv_time.get("HRV_RMSSD", np.nan).iloc[0]
}
print(features)
```

```python
mapping_df = pd.read_csv("./ecg-arrhythmia-1.0.0/ConditionNames_SNOMED-CT.csv")
```

```python
selected_codes = ["426177001", "426783006", "164889003", "427084000"]

filtered_df = mapping_df[mapping_df["Snomed_CT"].astype(str).isin(selected_codes)]

mapping_dict = dict(zip(filtered_df["Snomed_CT"].astype(str), filtered_df["Acronym Name"]))

print(mapping_dict)
```

```python
def get_arrhythmia_label(comments, mapping):
    for comment in comments:
        if comment.startswith("Dx:"):
            codes_str = comment[len("Dx:"):].strip()
            # Split by comma to get individual diagnosis codes
            codes = [code.strip() for code in codes_str.split(",")]
            # Check each code if it exists in the mapping dictionary
            for code in codes:
                # If a match is found return the corresponding label
                if code in mapping:
                    return mapping[code]
            # If no code from this comment matches, you might choose to return a default value or continue.
            return "Unknown"
    # If no "Dx:" field is found, return a default label.
    return "Unknown"
label = get_arrhythmia_label(record.comments, mapping_dict)
label
```
